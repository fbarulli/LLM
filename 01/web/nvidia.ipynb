```python
import os
import requests
import pandas as pd
from dotenv import load_dotenv
from IPython.display import display

# Load the environment variables
load_dotenv()
api_key = os.getenv("NVIDIA_API_KEY")

if not api_key:
    print("❌ NVIDIA_API_KEY not found in the environment. Please check your .env file.")
else:
    print(f"🔑 Authenticating with key: {api_key[:10]}...")
    
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Accept": "application/json"
    }

    # Query the models endpoint
    response = requests.get(
        url="https://integrate.api.nvidia.com/v1/models",
        headers=headers
    )

    if response.status_code == 200:
        models_data = response.json()
        print(f"✅ Success! Found {len(models_data['data'])} models.\n")
        
        # Parse the JSON data into a clean Pandas DataFrame
        models_list = []
        for model in models_data['data']:
            models_list.append({
                "Model ID": model.get("id"),
                "Type": model.get("object"),
                "Owned By": model.get("owned_by")
            })
        
        df = pd.DataFrame(models_list)
        
        # Sort by Model ID alphabetically for readability
        df = df.sort_values(by="Model ID").reset_index(drop=True)
        
        # Display the scrollable table in Jupyter
        display(df)
        
    else:
        print(f"❌ Failed with status code: {response.status_code}")
        print(response.text)
```
